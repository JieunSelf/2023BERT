```python
import pandas as pd
PATH = '../corpus/result.tsv'
df = pd.read_csv(PATH, encoding='utf-8', sep='\t')

df[10:30]
```

```python
df['age'].value_counts().sort_index()
```

```python
# label 생성
# 6~11세 : 0, 12~18세 : 1
df['label'] = [0 if s <11 else 1 for s in df['age']]
df[10:30]
```

```python
df.dtypes
```

```python
df.groupby(['label'], as_index=False).mean()
```

```python
df.groupby(['label'], as_index=False).std()
```

```python
df.groupby(['age'], as_index=False).mean()
```

```python
# t-test
from scipy.stats import levene
young = df['word_score'][df.label == 0]
old = df['word_score'][df.label == 1]
levene(young, old)
```

```python
# Welche's t-test
from scipy import stats
stats.ttest_ind(young, old, equal_var = False)
```

```python
# pvalue가 .05보다 작기 때문에 유의미한 차이가 있다고 볼 수 있다.
```
